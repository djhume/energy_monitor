```python
def weekly_stats():
    '''Function to look at the last week of demand'''
    def get_data():
        db = MySQLdb.connect (host = "192.168.1.120",user = "dave",passwd = "cc",db = "electricity")
        cc = sql.read_frame('SELECT time,power,temp FROM consumption WHERE time >= DATE_ADD(CURDATE(), INTERVAL -7 DAY);', db)
        cc.index = cc.time
        del cc['time']
        return cc
    
    return get_data()
    
def monthly_stats():
    '''Function to look at the last month of demand'''
    def get_data():
        db = MySQLdb.connect (host = "192.168.1.120",user = "dave",passwd = "cc",db = "electricity")
        cc = sql.read_frame('SELECT time,power,temp FROM consumption WHERE time BETWEEN SUBDATE(CURDATE(), INTERVAL 1 MONTH) AND NOW();', db)
        cc.index = cc.time
        del cc['time']
        return cc
    
    return get_data()

def so_far_this_month():
    '''Function to look at demand so far this month'''
    def get_data():
        db = MySQLdb.connect (host = "192.168.1.120",user = "dave",passwd = "cc",db = "electricity")
        cc = sql.read_frame('SELECT time,power,temp FROM consumption WHERE time >= CURDATE() - INTERVAL DAY(CURDATE())-1 DAY;', db)
        cc.index = cc.time
        del cc['time']
        return cc
    
    return get_data()

def last_month():
    '''Function to look at demand last month'''
    def get_data():
        db = MySQLdb.connect (host = "192.168.1.120",user = "dave",passwd = "cc",db = "electricity")
        cc = sql.read_frame('''SELECT time,power,temp FROM consumption WHERE time >= DATE_FORMAT(CURDATE() - INTERVAL 1 MONTH, '%Y/%m/01') 
                                                                         AND time <  DATE_FORMAT(CURDATE(), '%Y/%m/01' );''', db)
        cc.index = cc.time
        del cc['time']
        return cc
    
    return get_data()

def calc_daily_cost(df,cost_per_kwh):
    cost = ((df.power.groupby([lambda x: x.date(),lambda y: y.hour]).mean())/1000.0) * cost_per_kwh
    daily_cost = cost.groupby(level=0).sum()
    
    return daily_cost
    
def plot_cost_so_far_this_month(ckwh):
    thismonth = so_far_this_month()
    lastmonth = last_month()
    
    thismonth_kwh = thismonth.power.groupby([lambda x: x.date(),lambda y: y.hour]).mean().sum()
    thismonth_cost = calc_daily_cost(thismonth,ckwh)
    
    lastmonth_kwh = lastmonth.power.groupby([lambda x: x.date(),lambda y: y.hour]).mean().sum()
    lastmonth_cost = calc_daily_cost(lastmonth,ckwh)
    
    fig = plt.figure(1,figsize=[20,12])
    ax1 = fig.add_subplot(211)
    lastmonth_cost.plot(kind='bar',ax=ax1)
    ylabel('$')
    title('%.2f units consumed so far this month; costing $%.2f' % (thismonth_kwh/1000.0,thismonth_cost.sum()))
    suptitle('%.2f units consumed last month; costing $%.2f' % (lastmonth_kwh/1000.0,lastmonth_cost.sum()))

    ax2 = fig.add_subplot(212)
    thismonth_cost.plot(kind='bar',ax=ax2)
    ylabel('$')
```
